# Peptide preparation

Let's create the molecular system with help of MolSysMT

```python
import molsysmt as msm
```

```python
molecular_system = msm.build.build_peptide('NLYIQWLKDGGPSSGRPPPS')
```

```python
molecular_system = msm.build.add_missing_terminal_cappings(molecular_system,
                                                           N_terminal=None, C_terminal=None)
```

```python
molecular_system = msm.structure.center(molecular_system, selection='molecule_type=="peptide"')
```

Let's have some information about the system:

```python
msm.info(molecular_system, element = 'group')
```

We can also view the 3D representation of the molecule:

```python
view = msm.view(molecular_system, viewer='NGLView')
view.clear()
view.add_ball_and_stick()
view
```

Finally, let's save the molecule in a pdb file.

```python
msm.convert(molecular_system, to_form='trp_cage.pdb')
```
